```python
# Create Function for Recall MetaData Detail
import json

def recall_variable(meta_data,name):
  output_list = []
  print(f'Variable Name: {name}')
  print('\n')
  print(meta_data.loc[name,'Label'])
  print('\n')
  if meta_data.loc[name,'Value'] == np.nan:
    print('Value non-identify')
  elif isinstance(meta_data.loc[name,'Value'],dict):
    for key in list(meta_data.loc[name,'Value'].keys()):
      val = meta_data.loc[name,'Value'][key]
      output_list.append((key,val))
    return display(pd.DataFrame(output_list))
  else:
    print(meta_data.loc[name,'Value'])
  print('\n')
```

# School QQQ Dataset

## Prepare Metadata
`school_meta` is Meta Data for School Dataset

```python
# Import
school_df, school_meta = pyreadstat.read_sav(data_path + 'CY08MSP_SCH_QQQ.SAV')

# Variable Name and Label
name_label_dict = school_meta.column_names_to_labels
variable = school_meta.variable_value_labels

name_list = []
label_list = []
variable_list = []

for key in list(name_label_dict.keys()):
  name = key
  label = name_label_dict[key]
  name_list.append(name)
  label_list.append(label)

for key in name_list:
  try:
      # ถ้าสามารถ run ได้
      variable_bykey = variable[key]
      variable_list.append(variable_bykey)
  except:
      variable_list.append(None)

# Create Name and Label Dataset
meta_name = pd.DataFrame(name_list,columns=['Name'])
meta_value = pd.DataFrame(name_list,columns=['Name'])

meta_name['Label'] = label_list
meta_value['Value'] = variable_list

# Join meta_name and meta_value
school_meta = pd.merge(meta_name, meta_value, on='Name', how='left')
```

```python
school_meta = school_meta.set_index('Name')
school_meta['Name'] = name_list

school_meta = school_meta[['Name','Label','Value']]
```

```python
school_meta.info()
```

```python
school_meta.to_excel(data_path+'school_meta_2022.xlsx', index=False)
print('school_meta Wrirt and Upload to Drive')
```
